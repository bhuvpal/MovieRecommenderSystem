```python
movies = movies.merge(credits,on='title')
```

```python
movies.drop('movie_id',axis=1,inplace=True)
```

```python
movies = movies[['id','title','overview','genres','keywords','cast','crew']]
```

```python
import ast
```

```python
def change(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(change)
```

```python
movies['keywords'] = movies['keywords'].apply(change)
```

```python
movies['cast'].values[0]
```

```python
def change3(obj):
    l = []
    count = 0
    for i in ast.literal_eval(obj):
        if count!=3:
           l.append(i['name'])
           count+=1
        else:
            break
    return l
```

```python
movies['cast'] = movies['cast'].apply(change3)
```

```python
def change4(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
    return l
```

```python
movies['crew'] = movies['crew'].apply(change4)
```

```python
movies
```

```python
movies['overview'] = movies['overview'].str.split()
```

```python
movies
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(' ','') for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(' ','') for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(' ','') for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
movies
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.drop(columns=['genres','keywords','cast','crew','overview'],inplace=True)
```

```python
movies['tags'] = movies['tags'].str.join(' ')
```

```python
movies['tags'].values[0]
```

```python
new_df = movies[~movies['tags'].isnull()]
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(corp):
    l = []
    for i in corp.split():
        l.append(ps.stem(i))
    return ' '.join(l)
```

```python
stem('loving and fucking and sucking')
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
len(cv.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cs = cosine_similarity(vectors)
```

```python
cs
```

```python
def recommend(movie_name):
    index = new_df[new_df['title']=='Avatar'].index[0]
    distances = cs[index]
    recommend = sorted(list(enumerate(cs[index])),reverse=True,key=lambda x:x[1])[1:6]
    for i in recommend:
        print(new_df.iloc[i[0]].title)
    
```

```python
recommend('Avatar')
```
